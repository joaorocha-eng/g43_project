```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("g43_Space_Missions_v2.csv", sep=';', encoding='latin1', parse_dates=['transaction_date', 'created_date'], dayfirst=True)
```

```python
df_agency = df.loc[:, ['agency_id', 'agency_name']]
df_agency = df_agency.drop_duplicates(subset='agency_id', keep='first').copy()
df_agency.rename(columns={'agency_id': 'id', 'agency_name': 'name'}, inplace=True)
```

```python
df_mission = df.loc[:, ['mission_id', 'mission_title', 'mission_category', 'created_date']]
df_mission = df_mission.drop_duplicates(subset='mission_id', keep='first').copy()
df_mission.rename(columns={'mission_id': 'id', 'mission_title': 'name', 'mission_category': 'mission_type', 'created_date': 'start_date'}, inplace=True)
```

```python
df_astronaut = df.loc[:, ['astronaut_id', 'astronaut_info', 'agency_id']]
df_astronaut = df_astronaut.dropna(subset=['astronaut_id']).copy()
df_astronaut = df_astronaut.drop_duplicates(subset='astronaut_id', keep='first')
df_astronaut.rename(columns={'astronaut_id': 'id', 'astronaut_info': 'name'}, inplace=True)
```

```python
df_agencymission = df.loc[:, ['agency_id', 'mission_id', 'transaction_date', 'amount']]
df_agencymission = df_agencymission.dropna(subset=['transaction_date', 'amount']).copy()
df_agencymission = df_agencymission.drop_duplicates(subset=['agency_id', 'mission_id', 'transaction_date'], keep='first')
```

```python
engine = create_engine('sqlite:///data/space_missions.db')
```

```python
df_agency.to_sql('Agency', con=engine, if_exists='replace', index=False)
df_mission.to_sql('Mission', con=engine, if_exists='replace', index=False)
df_astronaut.to_sql('Astronaut', con=engine, if_exists='replace', index=False)
df_agencymission.to_sql('AgencyMission', con=engine, if_exists='replace', index=False)
```

```python
result = pd.read_sql('AgencyMission', con=engine)
```

```python
engine.dispose()
```
